```python
#Import packages 
import scipy.integrate
import numpy
import matplotlib.pyplot as plt
```

```python
def SIR_MODEL(y, t, beta, gamma):
    S, I, R = y
    dS_dt = -beta*S*I
    dI_dt = beta*S*I - gamma*I
    dR_dt = gamma*I

    return([dS_dt,dI_dt,dR_dt])
```

```python
#initial conditions

S0 = 0.9
I0 = 0.1
R0 = 0.0
beta = 0.3
gamma = 0.1
```
